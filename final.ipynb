#Recuento de evasión por variables categóricas

#Conteo de evasión por variables numéricas

```python
datos.groupby("Churn")[["tenure", "Monthly", "Total", "Cuentas_Diarias"]].describe()
```

```python
cols_numericas = ["tenure", "Monthly", "Total", "Cuentas_Diarias"]

for col in cols_numericas:
    datos[col] = pd.to_numeric(datos[col], errors="coerce")

```

```python
datos = datos.dropna(subset=cols_numericas)
```

```python
datos.groupby("Churn")[cols_numericas].mean()
```
